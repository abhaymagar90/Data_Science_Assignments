Q1. What is multiprocessing in python? Why is it useful?

Multiprocessing refers to the ability of a system to support more than one processor at the same time. Applications in a multiprocessing system are broken to smaller routines that run independently. The operating system allocates these threads to the processors improving performance of the system.

Q.2. What are the differences between multiprocessing and multithreading?

* Multithreading represents concurrency whereas multiprocessing represents parallelism.
* Python does not support multithreading in the case of parallel computing whereas Python supports multiprocessing in the case of parallel computing.
* In multithreading, multiple threads at the same time are generated by a single process whereas In multiprocessing, multiple threads at the same time run across multiple cores.
* Multiprocessing executes many processes simultaneously, whereas multithreading executes many threads simultaneously.
* Multiprocessing uses two or more CPUs to increase computing power, whereas multithreading uses a single process with multiple code segments to increase computing power.

Q3. Write a python code to create a process using the multiprocessing module.

```python
import multiprocessing, time

def sqr(x):
   print(x**2)

if __name__ == "__main__":
  print("This is multiprocessing")

  for x in range(1,5+1):
    m1 = multiprocessing.Process(target = sqr, args=(x,))
    m1.start()
    time.sleep(1)
```

Q4. What is a multiprocessing pool in python? Why is it used?

The Python Multiprocessing Pool class allows you to create and manage process pools in Python. Here, we import the Pool class from the multiprocessing module. In the main function, we create an object of the Pool class. The pool.map() takes the function that we want parallelize and an iterable as the arguments. It runs the given function on every item of the iterable. 

Q5. How can we create a pool of worker processes in python using the multiprocessing module?

```python
import multiprocessing, time

def sqr(x):
  return(x**2)

pool = multiprocessing.Pool(processes = 5)
op = pool.map(sqr, [5,3,8,6,9])
print(op)
```

Q6. Write a python program to create 4 processes, each process should print a different number using the
multiprocessing module in python.

```python
import multiprocessing, time
n1 = 1
n2 = 2
n3 = 3
n4 = 4

m1 = multiprocessing.Process(target = print(n1))
m2 = multiprocessing.Process(target = print(n2))
m3 = multiprocessing.Process(target = print(n3))
m4 = multiprocessing.Process(target = print(n4))

m_list = [m1,m2,m3,m4]

for i in m_list:
  i.start()
  time.sleep(0.5)
  i.join()
```
