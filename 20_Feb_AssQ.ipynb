Q1. Explain GET and POST methods.

GET and POST methods are used to input data from the client side to the server side so that result is displayed over the webpage depending upon these inputs.
* GET is less secure. Here data is appended into the URL itself. Hence, data is available publicly.

* POST is more secure as data is inputted using forms. Hence, data is not available publicly.


Q2. Why is request used in Flask?

The main aim of request is to accept the incoming data from the client so as to perform operations/ function with the data.

request is used with GET and POST
* GET  ->  request.args.get()
* POST  -> request.form[]

Q3. Why is redirect() used in Flask?

Flask redirect is defined as a function or utility in Flask which allows developers to redirect users to a specified URL.

Q4. What are templates in Flask? Why is the render_template() function used?

Templates are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates. We use templates to render HTML which will display in the user's browser.

render_template() is used to serve an HTML template as the response. Using it, an html file can be displayed on the web page
