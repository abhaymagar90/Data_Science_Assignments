Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

* Scatterplot -> It can be used to show relationship between two columns
* Lineplot -> more commonly used to keep track of changes over time
* Histogram -> To find frequency of different data points of a column
* Barplot -> Mostly used to find relationship between a numeric and categorical variable
* Boxplot -> They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in the dataset.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x = fmri.timepoint, y=fmri.signal)
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x=titanic.pclass, y = titanic.age)
```

```python
sns.boxplot(x=titanic.pclass, y = titanic.fare)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamond = sns.load_dataset('diamonds')
diamond
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18,4))
sns.histplot(ax = ax, data = diamond, x = diamond['price'], hue='cut')
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue = 'species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
flights
```

```python
df_values = flights.pivot("year", "month", "passengers")
```

```python
sns.heatmap(df_values, annot=True, fmt="d")
```
