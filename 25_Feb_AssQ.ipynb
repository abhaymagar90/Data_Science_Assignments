Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

```python
d01 = input("enter date 1 ")
d1 = pd.to_datetime(d01)
print(d1)
d02 = input("enter date 2 ")
d2 = pd.to_datetime(d02)
print(d2)

rounded_dt1 = d1.replace(hour=0, minute=0, second=0, microsecond=0)
rounded_dt2 = d2.replace(hour=0, minute=0, second=0, microsecond=0)
delta = (rounded_dt2 - rounded_dt1)
print(delta.days)
```

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

```python
import csv

import pandas as pd
filepath = input("Enter filepath name: ")
df = pd.read_csv(filepath)
df.head()
```

```python
col = input("Enter column name ")
df[col].dtypes
```

```python
df[col] = df[col].astype('category')
```

```python
df[col].dtypes
```

```python
df[col]
```

```python
ord = []
for i in range(len(pd.unique(df[col]))):
  m = input("enter category")
  ord.append(m)
ord
```

```python
df[col] = df[col].cat.set_categories(ord, ordered=True)
```

```python
df[col]
```
