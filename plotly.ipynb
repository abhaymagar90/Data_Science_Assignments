Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
from re import X
import plotly.express as px

px.scatter(x = titanic['age'], y=titanic['fare'], title="Age vs Fare").update_layout(xaxis_title="Age", yaxis_title="Fare")
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

We use a distplot (also known as a distribution plot) to represent data in histogram form. It is a univariant set of collected data, which means the data distribution of one variable will be shown against another variable.

```python
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.rand(100)-1
x2 = np.random.rand(100)
x3 = np.random.rand(100)+1

x = [x1,x2,x3]
grp = ["Group1","Group2", "Group3"]

fig = ff.create_distplot(x,grp, show_hist=False)
fig.update_layout(title_text='Distplot')
fig.show()
```
