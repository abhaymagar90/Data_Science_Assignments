Q.3] What errors are defined in the ArithmeticError class? Explain any two with an example.

Arithmetic class has two types of errors:

* ZeroDivisionError
* OverflowError

```python
#ZeroDivisionError

try:
  10/0
except ZeroDivisionError as e:
  print(e)
```

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except OverflowError as e:
  print(e)
```

Q.4] Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

You can use LookupError exception class to handle both IndexError and KeyError exception classes.

```python
#keyerror

try:
  d = {"Name":"Abhay Magar", "Address":"India"}
  d["Hobbies"]
except KeyError as e:
  print(e)
```

```python
#IndexError

try:
  a = [1,2,3,4,5]
  print(a[2])
  print(a[20])
except IndexError as e:
  print(e)
```

Q.5] Explain ImportError. What is ModuleNotFoundError?

ModuleNotFoundError is a kind of ImportError. The ImportError is raised when an import statement has trouble successfully importing the specified module. Typically, such a problem is due to an invalid or incorrect path.

```python
try:
  import abhay
except ImportError as e:
  print(e)
```

Q.6] List down some best practices for exception handling in python.

```python
# Always use specific exception and not generic exception.   Exception(Generic)         ZeroDivisionError/ TypeError/ NameError(Specific)

try:
  10/0
except Exception as e:           # This is not a good practice
  print(e)
```

```python
try:
  10/0
except ZeroDivisionError as e:          # This is good practice, as we specify the error type
  print(e)
```

```python
# Always print a valid and proper message while handling, which helps to understand exception better and makes it easy to debug

try:
  a = [1,2,3,4,5]
  a[10]
except IndexError as e:
  print("Error : ", e, "\nYou can specify maximum index of ", len(a)-1)
```

```python
# Always try to log

import logging
logging.basicConfig(filename = "error.log",force = True, level = logging.INFO )


try:
  logging.info("This is start of program")
  10 / 0
except ZeroDivisionError as e:
  logging.error(e)
```

```python
# Always try to avoid multiple exception handling

try:
  10/0
except FileNotFoundError as e:
  print(e)
except IndexError as e:
  print(e)
except NameError as e:
  print(e)
except ZeroDivisionError as e:
  print(e)

# Here, the only possible exception for code in try block is ZeroDivisionError. Hence, no need to introduce other handling methds
```
